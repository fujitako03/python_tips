```python
import pandas as pd 
import plotly.express as px
```

```python
# preprocess dataset
df = pd.read_csv("https://covid19.who.int/WHO-COVID-19-global-data.csv",encoding="utf-8")
df = df[(df["Country"] == "France") | (df["Country"] == "Brazil")]
df["Date_reported"] = pd.to_datetime(df["Date_reported"])
df["month"] = df["Date_reported"].dt.month
```

```python
df.head()
```

### 折れ線グラフ

```python
fig = px.line(df, x="Date_reported", y="New_cases", color='Country')
fig.update_layout(
    title='国別 新規感染者数推移',
    xaxis_title='新規感染者数',
    yaxis_title='日付',
    xaxis_tickformat="%Y-%m", # x軸_表示形式
    xaxis_dtick="M2", # x軸_表示間隔（M1,M2,D,7*86400000.0)
    yaxis_tickformat = ',', # y_軸表示形式
    yaxis_dtick=20000, # y軸_表示間隔
    yaxis_ticksuffix = '人', # y軸_単位
    )
fig.update_layout()
fig.show()
```

 ### 棒グラフ

```python
# 前処理（これをしないと細かい積み上げになる）
df_plot = df.groupby("month", as_index=False).sum()

# plot
fig = px.bar(df_plot, x="month", y="New_cases")
fig.update_layout(
    title='国別 新規感染者数推移',
    xaxis_title='月',
    yaxis_title='新規感染者数',
    # xaxis_tickformat="%Y-%m", # x軸_表示形式
    xaxis_tickvals=df['month'], 
    yaxis_tickformat = ',', # y_軸表示形式
    yaxis_dtick=500000, # y軸_表示間隔
    yaxis_ticksuffix = '人', # y軸_単位
    )
fig.show()
```

# パラメータ設定例
### dtick
- 1ヶ月ごと：'M1'
- 2ヶ月ごと: 'M2'で2ヶ月ごとにラベル表示
- 1日ごと: 'D'で１日ごとにラベル表示
- 7日ごと: 7*86400000.0 

### tickformat
- 2018-05-01："%Y-%m-%d"
- 5/1："%_m/%-d"
- 5/1(Tue)："%_m/%-d(%a)"
